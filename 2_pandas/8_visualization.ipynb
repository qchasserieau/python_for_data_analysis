Plotting and Visualization

```python
import pandas as pd
import numpy as np
%Â matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
df = pd.DataFrame(list(range(20)))
df.columns = ['linear']
```

```python
df.head()
```

```python
df['square'] = np.power(df['linear'], 2)
df['tenfold'] = df['linear'] * 10
```

```python
import random
```

```python
random.random()
```

```python
help(random.random)
```

```python
df['random'] = df['square'].apply(lambda s: (0.75 + random.random()/2) *s)
df['randint'] = df['linear'].apply(lambda s: 5* random.randint(0, int(s/2)))
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plot = df.plot()
```

```python
fig = plot.get_figure()
fig.savefig(r'../plot/plot.png')
```

```python
df.hist(bins=20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [5, 5]
df.plot(kind='scatter', x='square', y='random')
```

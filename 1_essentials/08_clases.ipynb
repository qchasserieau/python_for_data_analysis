# Clases
En python, todo es un objeto, las variables, las listas, los dicciónarios : todo

Pueden crear sus proprios objetos : las clases

En realidad : definimos las probriedades del objeto con la clase

Y cuando instanciamos la clase : creamos un objeto cuyo tipo es la misma clase

Las clases se declaran así:
```python
class ClassName:
    methods (functions)```


````python
class FirstClass:
    "This is an empty class"
    pass
help(FirstClass)
````

````python
test = FirstClass()
type(test)
````

Las clases pueden contenir datos, que llamamos atributos

````python
class myClass():
    my_attr = 1
    
a = myClass()
a.my_attr
````

Las clases pueden tener funciones proprias que llamamos méthodos

````python
class myClass():
    my_attribute = 5
    def add(self, x):
        return x + self.my_attribute

a = myClass()
a.add(3)    
````

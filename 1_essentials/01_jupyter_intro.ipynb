# Modes commande et édition
Le notebook est utilisable selon deux modes:
* le mode commande: on y accède en tapant `echap`. Il permet de naviguer facilement entre les cellules, avec les flèches directionnelles du clavier, et d'accéder à certains raccourcis
* le mode édition: on y accède en tapant `entrée`. Il permet d'éditer le contenu des ceulules.

# Le kernel est l'entité sur laquelle le code s'exécute

Vous pouvez l'arrêter et le redémarrer.

C'est particulièrement pratique si:
* vous avez lancé un calcul long que vous voulez annuler
* vous souhaitez vider la mémoire pour éliminer les variables que vous avez pu instancier, et repartir d'un état "propre"

Essayez de couperr l'exécution du kernel en appuyant sur le bouton <button class='btn btn-default btn-xs'><i class='icon-stop fa fa-stop'></i></button> dans la barre des tâches juste après avoir lancé l'exécution de cette cellule:

```python
import time
time.sleep(10)
```

```python
print(a)
```

Essayez de redémarrer le kernel en appuyant sur le bouton <button class="btn btn-default btn-xs"><i class="fa-repeat fa"></i></button> et de sortir à nouveau la valeur de la variable a:

# Outputs
Python prévoit deux sorties pour les choses que l'on imprime: 
* stdout: la sortie standard pour imprimer des choses noramles
* std err: la sortie standard pour imprimer les erreurs et avertissements

```python
print("hi, stdout")
```

```python
import sys
print('hi, stderr', file=sys.stderr)
```

```python
#stderr est utilisée aussi pour les alertes (warnings)
import warnings
warnings.warn('hello')
```

# L'output est asynchrone
Ca veut dire que l'on va imprimer dans le notebook les sorties dès qu'elles sont disponibles

```python
import time
for i in range(5):
    time.sleep(1)
    print(i)
```

# L'output peut être réduit (collapsed)
Lorsque la sortie prend trop de place, il est possible de cliquer à gauche de celle-ci pour réduire son espace.

```python
for i in range(50):
    print(i)
```

Lorsque la zone de sortie est trop grande, elle est réduite par défaut

```python
for i in range(1000):
    print(2**i)
```

Lorsque vous avez lancé le notebook, une fenêtre noire s'est ouverte. Ce qu'il s'est passé:
